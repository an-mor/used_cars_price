Есть неявлые дубликаты в названиях производителей - например, mercedes и mercedes-b. Создаю функцию для обработки таких дубликатов.

```python
# функция для обработки названий производителей
def make_edit_name(maker):
    if 'dodge' in maker:
        maker = 'dodge'
    elif 'ford' in maker:
        maker = 'ford'
    elif 'gmc' in maker:
        maker = 'gmc'
    elif 'rover' in maker and 'land' in maker:
        maker = 'landrover'
    elif 'mazda' in maker:
        maker = 'mazda'
    elif 'mercedes' in maker:
        maker = 'mercedes'
    elif 'hyundai' in maker:
        maker = 'hyundai'
    elif maker == 'vw':
        maker = 'volkswagen'
    return maker
```

```python
# функция для обработки типов кузова
def body_edit_type(body):
    if 'coupe' in body or body == 'koup':
        return('coupe')
    if 'convertible' in body:
        return('convertible')
    if 'wagon' in body:
        return('wagon')
    if 'van' in body and body != 'minivan':
        return('van')
    if 'sedan' in body:
        return('sedan')
    return body
```

### Catboost cv

Обучаю модель Catboost с кросс-валидацией. Убраны лишние попытки с другими значениями параметров.

```python
train_pool = Pool(features, target, cat_features=cat_cols)
```

```python
params = {
    'iterations': 2000,
    'learning_rate': 0.14,
    'loss_function': 'RMSE',
    'eval_metric': 'MAPE',
    'random_seed': RND_STATE
}
```
